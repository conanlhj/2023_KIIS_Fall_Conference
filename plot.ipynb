```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 정의
datasets = ['Constant', 'LDP', 'LDP+CC']
loss = [0.7258, 0.4301, 0.2768]
accuracy = [0.6920, 0.9040, 0.9440]

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(10, 6))

# 왼쪽 y축 (Loss)
ax1.set_xlabel('Dataset')
ax1.set_ylabel('Loss', color='blue')
ax1.bar(datasets, loss, color='blue', alpha=0.6, label='Loss')
ax1.tick_params(axis='y', labelcolor='blue')

# 오른쪽 y축 (Accuracy)
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', color='red')
ax2.plot(datasets, accuracy, color='red', marker='o', label='Accuracy')
ax2.tick_params(axis='y', labelcolor='red')

# 그래프 표시
fig.tight_layout()/
plt.title('Comparison of Loss and Accuracy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 정의
datasets = ['Constant', 'LDP', 'LDP+CC']
loss = [0.7258, 0.4301, 0.2768]
accuracy = [0.6920, 0.9040, 0.9440]

# 데이터를 약간의 offset을 주어 그래프에 겹치지 않게 함
x = np.arange(len(datasets))
width = 0.25  # 막대 너비 조절

# 그래프 생성
fig, ax1 = plt.subplots(figsize=(16, 7))

# 글씨 크기 및 스타일 설정
plt.rcParams.update({
    'font.size': 25,
    'font.family': 'Times New Roman'
})

# 왼쪽 y축 (Loss)
ax1.set_xlabel('Dataset')
ax1.set_ylabel('Loss', color='blue')
ax1.bar(x - width/2, loss, width, color='blue', alpha=0.6, label='Loss')
ax1.tick_params(axis='y', labelcolor='blue', size=14)
ax1.set_xticks(x)
ax1.set_xticklabels(datasets)

# 오른쪽 y축 (Accuracy)
ax2 = ax1.twinx()
ax2.set_ylabel('Accuracy', color='red')
ax2.bar(x + width/2, accuracy, width, color='red', alpha=0.6, label='Accuracy')
ax2.tick_params(axis='y', labelcolor='red', size=14)

# 그래프 표시
fig.tight_layout()
plt.title('Comparison of Loss and Accuracy', fontsize=30)
plt.show()
```
